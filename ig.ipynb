```python
def scrapear(usuario, cantidad_usuarios_a_scrapear = 1000000, scrapear_seguidores = False, scrapear_seguidos = False, download = True):
    
    url_usuario = 'https://www.instagram.com/'+usuario

    # SEGUIDORES
    if scrapear_seguidores == True:
        
        print(f'Scrapeando seguidores de {usuario}\n\n')

        # Entrar al usuario
        start = time.time()
        driver.get(url_usuario)
        sleep(8)

        # Acceder a seguidores
        seguidores_boton = driver.find_element('xpath',"//ul[@class='_aa_7']/li[2]//span")
        seguidores_boton.click()
        sleep(3)

        # cantidad de seguidores
        can_seguidores = driver.find_element('xpath',"//ul[@class='_aa_7']/li[2]//span").text
        if '.' in can_seguidores:
            cantidad_decimal = len(can_seguidores.split('.')[-1][:-1])
        else:
            cantidad_decimal = 0
        can_seguidores = int(can_seguidores.replace('K','0'*(3-cantidad_decimal)).replace('M','0'*(6-cantidad_decimal)).replace('.','').replace(',',''))

        # cuanto hay que scrollear
        scrollear_inicial = round(can_seguidores/12)+1

        # instanciamos seguidores_total
        seguidores_total = []
        cantidad_encontrados = []

        # scrollear y apendear
        for i in range(scrollear_inicial):
            i += 1

            #Cada 40 scrolls, pausa de 8 minutos
            if i % 40 == 0:
                minutos = random.uniform(3, 7)
                print(f'\n\n\nHaciendo pausa de {round(minutos)} minutos\n\n\n')
                time.sleep(minutos*60)

            # Print
            print(f'Scrolleando {i} de {scrollear_inicial}')

            # Scroll down
            driver.execute_script("""
                var scrollt = document.querySelector('div[role="dialog"] ._aano')
                scrollt.scrollTop = scrollt.scrollHeight
                """)

            time.sleep(random.uniform(8, 12))

            # Nuevos seguidores
            # Español
            seguidores = driver.find_elements('xpath',f"(//div[@class='_aae-']/li)[position()>={len(seguidores_total)}]//*/a")
            seguidores = [u.text for u in seguidores if len(u.text)>0]
            if len(seguidores) == 0:
                #Ingles
                seguidores = driver.find_elements('xpath',f"//div[position()>={len(seguidores_total)}]//*/span/a[@role = 'link']")
                seguidores = [elem.get_attribute('href').split('/')[-2] for elem in seguidores]

            # agregamos los nuevos seguidores
            seguidores_total.extend(seguidores)
            print(f'Hasta el momento, se han encontrado {len(seguidores_total)} seguidores de {usuario}\n')

            # Si no encontro nada en las ultimas 3 iters, break
            cantidad_encontrados.append(len(seguidores_total))
            try:
                if cantidad_encontrados[-1] == cantidad_encontrados[-2] and cantidad_encontrados[-1] == cantidad_encontrados[-3]:
                    print('\n\n\nINTERRUPCION: NO SE ENCUENTRAN NUEVAS CUENTAS\n\n\n')
                    end = time.time()
                    print(f'\n\nEsto tomó {round((end-start)/60)} minutos')
                    driver.close()
                    break
            except:
                pass

            # break si se alcanzo maximo de usuarios
            if len(seguidores_total) > cantidad_usuarios_a_scrapear:
                break
                    
                time.sleep(random.uniform(8, 12))
    
        if download == True:
            descargar(usuario, seguidores_total, 'seguidores')

    # SEGUIDOS
    if scrapear_seguidos == True:
        
        print(f'Scrapeando seguidos de {usuario}\n\n')
        
        # Entrar al usuario
        start = time.time()
        driver.get(url_usuario)
        sleep(8)

        # Acceder a seguidos
        seguidos_boton = driver.find_element('xpath',"//ul[@class='_aa_7']/li[3]//span")
        seguidos_boton.click()
        sleep(3)

        # cantidad de seguidos
        can_seguidos = driver.find_element('xpath',"//ul[@class='_aa_7']/li[3]//span").text
        if '.' in can_seguidos:
            cantidad_decimal = len(can_seguidos.split('.')[-1][:-1])
        else:
            cantidad_decimal = 0
        can_seguidos = int(can_seguidos.replace('K','0'*(3-cantidad_decimal)).replace('M','0'*(6-cantidad_decimal)).replace('.','').replace(',',''))

        # cuanto hay que scrollear
        scrollear_inicial = round(can_seguidos/12)+1

        # instanciamos seguidos_total
        seguidos_total = []
        cantidad_encontrados = []

        # scrollear y apendear
        for i in range(scrollear_inicial):
            i += 1
            
            # Cada 40 scrolls, pausa de 8 minutos
            if i % 40 == 0:
                minutos = random.uniform(3, 7)
                print(f'\n\n\nHaciendo pausa de {round(minutos)} minutos\n\n\n')
                time.sleep(minutos*60)

            # Print
            print(f'scrolleando {i} de {scrollear_inicial}')

            # Scroll down
            driver.execute_script("""
                var scrollt = document.querySelector('div[role="dialog"] ._aano')
                scrollt.scrollTop = scrollt.scrollHeight
                """)
            time.sleep(random.uniform(8, 12))

            # Nuevos seguidos
            # Español
            seguidos = driver.find_elements('xpath',f"(//div[@class='_aae-']/li)[position()>={len(seguidos_total)}]//*/a")
            seguidos = [u.text for u in seguidos if len(u.text)>0]
            if len(seguidos) == 0:
                #Ingles
                seguidos = driver.find_elements('xpath',f"//div[position()>={len(seguidos_total)}]//*/span/a[@role = 'link']")
                seguidos = [elem.get_attribute('href').split('/')[-2] for elem in seguidos]

            # agregamos los nuevos seguidos
            seguidos_total.extend(seguidos)
            print(f'Hasta el momento, se han encontrado {len(seguidos_total)} seguidos de {usuario}\n')

            # Si no encontro nada en las ultimas 3 iters, break
            cantidad_encontrados.append(len(seguidos_total))
            try:
                if cantidad_encontrados[-1] == cantidad_encontrados[-2] and cantidad_encontrados[-1] == cantidad_encontrados[-3]:
                    print('\n\n\nINTERRUPCION: NO SE ENCUENTRAN NUEVAS CUENTAS\n\n\n')
                    end = time.time()
                    print(f'\n\nEsto tomó {round((end-start)/60)} minutos')
                    break
            except:
                pass

            # break si se alcanzo maximo de usuarios
            if len(seguidos_total) > cantidad_usuarios_a_scrapear:
                break

            time.sleep(random.uniform(8, 12))
        
        if download == True:
            descargar(usuario, seguidos_total, 'seguidos')
    
    
    # Return
    if scrapear_seguidores == True and scrapear_seguidos == True:
        return seguidores_total, seguidos_total
    elif scrapear_seguidores == True and scrapear_seguidos == False:
        return seguidores_total
    elif scrapear_seguidores == False and scrapear_seguidos == True:
        return seguidos_total 
    else:
        return print('Por favor, instanciar la funcion con scrapear_seguidores o scrapear_seguidos en True')
```
