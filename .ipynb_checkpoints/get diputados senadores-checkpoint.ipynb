```python
def get_match(string, lista_strings):
    string = unidecode.unidecode(string.strip().lower())
    string = re.findall('^[a-zA-Z0-9\s]*',string)[0]
    lista_strings = unidecode.unidecode(string.strip().lower())
    lista_strings = re.findall('^[a-zA-Z0-9\s]*',string)
    mejor_match, puntaje = process.extractBests(string, lista_strings, scorer=fuzz.token_set_ratio)[0]
    return puntaje
```

```python
def insta_bio_search(link):
    key_words = ['senado','diputad','politic','medic','abogad'] #if one of this match in bio, keep ig
    user = link.split('/')[-2]
    try:
        L
    except:
        L = instaloader.Instaloader()
    profile = instaloader.Profile.from_username(L.context, user)
    bio = profile.biography
    bio = unidecode.unidecode(bio.strip().lower())
    for k in key_words:
        if k in bio:
            return True
            break
    return False
```

```python
# iterar y buscar instagrams
def google_search_ig(df, cargo, punto_de_corte):
    
    #iterar por los politicos sin instagram
    for i_df, name_orginal in enumerate(df['nombres'].values):
        if df['igs'][i_df] == '':
            
            # Creamos el nombre original y el buscable en google
            name_orginal = unidecode.unidecode(name_orginal.strip())
            name_buscable = cargo+'+'+'+'.join(name_orginal.split(' '))

            # Buscamos y guardamos los links
            driver.get(f'https://www.google.com/search?q={name_buscable}+Instagram')
            sleep(4)
            links = driver.find_elements('xpath','//*[@id="rso"]//*/div/div[1]/div[1]/a')

            # Iteramos y nos quedamos con el instagram que mejor se ajuste al nombre y su
            found = ''
            puntaje = 0
            for i,x in enumerate(links):
                x_link = x.get_attribute('href')
                if 'instagram' in x_link:
                    new_puntaje = get_match(name_orginal, [x.text])
                    if new_puntaje > puntaje and new_puntaje > punto_de_corte:
                        try:
                            if insta_bio_search(x_link):
                                puntaje = new_puntaje
                                found = x_link
                        except:
                            pass

            df['igs'][i_df] = found

            print(i_df)
            print(name_orginal)
            print(found)
            print('\n')
    print(f'Encontrados {len(df[df['igs'] != ''])} de {len(df)}.')
            
```

```python
#SENADORES
# ir a la pagina con senadores
driver.get('https://www.senado.gob.ar/senadores/listados/listaSenadoRes')
sleep(5)

# Nombres
nombres = driver.find_elements('xpath','//*[@id="senadoresTabla"]/tbody/tr/td[2]/a')
nombres = [t.text for t in nombres]
nombres = [n.replace(',','') for n in nombres]

# Partido
partidos = driver.find_elements('xpath','//*[@id="senadoresTabla"]/tbody/tr/td[4]')
partidos = [t.text for t in partidos]

# Distritos
distritos = driver.find_elements('xpath','//*[@id="senadoresTabla"]/tbody/tr/td[3]')
distritos = [t.text for t in distritos]

# Instagrams
igs = []
for i in range(1,len(nombres)+1):
    ig = driver.find_elements('xpath',f'//*[@id="senadoresTabla"]/tbody/tr[{i}]/td[6]/ul/li[4]/a')
    ig = [i.get_attribute('href') for i in ig]
    ig = [i for i in ig if 'instagram' in i]
    ig = [ig[0] if len(ig)>0 else ''][0]
    igs.append(ig)

df_senadores = pd.DataFrame([nombres,igs,partidos,distritos]).T
df_senadores.columns = ['nombres','igs','partidos','distritos']

google_search_ig(df_senadores, 'senador', 70)
```

```python
#DIPUTADOS
# ir a la pagina con diputados
driver.get('https://www.hcdn.gob.ar/diputados/listadip.html')
sleep(5)


# Nombres
nombres = driver.find_elements('xpath','//*[@id="tablaDiputados"]/tbody/tr/td[2]/a')
nombres = [t.text for t in nombres]
nombres = [n.replace(',','') for n in nombres]

# Partido
partidos = driver.find_elements('xpath','//*[@id="tablaDiputados"]/tbody/tr/td[6]')
partidos = [t.text for t in partidos]

# Distritos
distritos = driver.find_elements('xpath','//*[@id="tablaDiputados"]/tbody/tr/td[3]')
distritos = [t.text for t in distritos]

# Instagrams
igs = [''] * len(nombres)

df_diputados = pd.DataFrame([nombres,igs,partidos,distritos]).T
df_diputados.columns = ['nombres','igs','partidos','distritos']
```

```python
google_search_ig(df_diputados, 'diputad', 70)
```
